```python
wordlist = []

# opens .txt file of top 1000 most used words
filefxn = open('1-1000.txt') 
# appends all words to list
wordlist.append(filefxn.read().splitlines())
filefxn.close()

# turns list of lists to list
wordlist = wordlist[0]
```

```python
for x in wordlist:
    if len(x) >= 5:
        print(x)
```

```python
modified_word_list = open('modified 1-1000.txt','w')

for x in wordlist:
    if len(x) >= 5: # grabs all words from list with 5 letters or more
        modified_word_list.write(x + "\n")

# 521 words grabbed out of the original 1000

# note this creates an empty line at the end of the file
# copy-pasted results into dictionary.txt file

modified_word_list.close() 
```
